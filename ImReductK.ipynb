```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import glob
import os
#import astropy\n",
from astropy.io import fits
from scipy.ndimage import interpolation, filters, morphology
import time
import image_registration
#from __future__ import division, print_function
import math
#from numpy.fft import fft2, ifft2, fftshift
#import scipy.ndimage.interpolation as ndii

#rename for proper sorting
imnames = []
imnames = glob.glob('I*.fits')
for i in range(len(imnames)):
    if len(imnames[i]) == 19:
        os.rename(imnames[i], imnames[i][0:11]+'0'+imnames[i][11:13]+'0'+imnames[i][13:])
        imnames[i] = imnames[i][0:11]+'0'+imnames[i][11:13]+'0'+imnames[i][13:]
    elif len(imnames[i]) == 20:
        if imnames[i][12] == '_':
            os.rename(imnames[i], imnames[i][0:11]+'0'+imnames[i][11:])
            imnames[i] = imnames[i][0:11]+'0'+imnames[i][11:]
        elif imnames[i][13] == '_':
            os.rename(imnames[i], imnames[i][0:14]+'0'+imnames[i][14:])
            imnames[i] = imnames[i][0:14]+'0'+imnames[i][14:]
    
imnames = sorted(imnames[:])
    
leak = fits.open("L_1_215_23_17_11.fits")
flat = fits.open("F_1_215_17_27_25.fits")
dark = fits.open("D_1_215_17_2_10.fits")

#make flatfield: gaussian smooth, then divide image by gaussian smoothed image
      
gflat = scipy.ndimage.gaussian_filter(flat[0].data - dark[0].data, 8)
      
flat = (flat[0].data - dark[0].data) / gflat
hdu6 = fits.PrimaryHDU(flat)
hdu6.writeto('Kflat.fits', clobber=True)
      
#align and subtract leak image

#threshhold image to find Sun center


#plt.plot(leak[0].data[0])
#plt.show()

#plt.imshow(morphology.morphological_laplace(np.log10(leak[0].data[0]), size=np.shape(leak[0].data[0]), mode='constant'))
#plt.colorbar()
#plt.show()

#a = np.zeros((4,1024,1024))
#a = leak[0].data > 5000 #nominal boundary for Sun (although somewhat arbitrary because pix vals increase toward center)

#leak_cent = [np.unravel_index(np.argmax(morphology.distance_transform_edt(a[i])),(1024,1024)) for i in range(4)]

#print leak_cent #tuple: (y,x) in ds9 image
#b = morphology.morphological_gradient(leak[0].data[0], np.shape(leak[0].data[0]))
#plt.imshow(b)
#plt.show()

leakshift = np.zeros((4,1024,1024))
leakshift[0] = interpolation.shift(leak[0].data[0], (1.5,-3)) # tuple: (y,x) in ds9 image
leakshift[1] = interpolation.shift(leak[0].data[1], (1.5,-3)) 
leakshift[2] = interpolation.shift(leak[0].data[2], (-0.5,0)) # tuple: (y,x) in ds9 image
leakshift[3] = interpolation.shift(leak[0].data[3], (-0.5,0)) 
hdu5 = fits.PrimaryHDU(leakshift)
hdu5.writeto('Kleakshift.fits', clobber=True)

imnames = []
imnames = sorted(glob.glob("I_1*.fits"))
print imnames[0:3]
#import fits image names\n",
N = len(imnames)
imlist = np.zeros((N,1024,1024))
print N #number of fits images
#for i in range(N):\n",
    #print fits.open(imnames[i])[0].data[0]\n",
    #imlist[i] = fits.open(imnames[i])[0].data[0]\n",
      
Kimnames = imnames
#print Kimnames

t = time.time()
from time import localtime, strftime
tstart = strftime("%a, %d %b %Y %H:%M:%S", localtime())
print tstart


reference = fits.open('20140804_210045hmiM.fits')
print reference.info()
reference.verify('fix')
print np.shape(reference[1].data)
reference = interpolation.zoom(reference[1].data, 0.25)


def findcenter(image, ref): #reference: HMI image
    shape = image.shape
    plt.imshow(image)
    plt.show()
    plt.imshow(ref)
    plt.show()
    #Pad images
    
    #image = np.pad(image, (512,),'edge')
    #ref = np.pad(ref, (512,),'edge')
    
    #Compute FFT of images
    f0 = np.fft.fftshift(np.fft.fft2(morphology.morphological_gradient(image,shape)/np.max(image)))
    f1 = np.fft.fftshift(np.fft.fft2(morphology.morphological_gradient(ref,shape)/np.max(ref)))
    print f0.shape, f0[0]
    #check for nans
    print np.isnan(np.any(f0)), np.isnan(np.any(f1))
    
    #Gaussian filter
    dnu = 1./(len(f0[0]))
    print dnu
    #Make frequency plane
    freq = np.fft.fftshift(np.fft.fftfreq(1024))
    print freq
    #Gfilter = np.zeros_like(f0)
    #mu = 500 # based one visual inspection of function--how to do this smarter?
    #sig = 10
    #print len(freq), len(Gfilter)
    #for i in range(len(freq)):
    #    Gfilter[i,j] = (np.exp(-(freq[i]-mu)**2./(2.*sig**2.))+np.exp(-(freq[i]-((len(f0)*dnu)-mu))**2/(2*sig**2)))
    #f0 = f0 * Gfilter
    #f1 = f1 * Gfilter

    #Achieve subpixel precision with Gaussian filter
    
    plt.plot(freq,np.log10(abs(f0[0])))
    plt.plot(freq,np.log10(abs(f0[225])))
    plt.plot(freq,np.log10(abs(f0[512])))
    plt.plot(freq,np.log10(abs(f0[750])))
    #plt.xlim(-0.05,0.05)
    #plt.ylim(0,1000)
    plt.xlabel("Frequency[pixel^-1]")
    plt.title("f0")
    plt.show()
    print np.unravel_index(np.argmax(abs(f0)), np.shape(f0))
    plt.plot(freq,np.log10(abs(f1[0])), label="0")
    plt.plot(freq,np.log10(abs(f1[225])), label="225")
    plt.plot(freq,np.log10(abs(f1[512])), label="512")
    plt.plot(freq,np.log10(abs(f1[750])), label="750")
    plt.legend()
    #plt.xlim(0,0.2)
    #plt.ylim(0,1000)
    plt.xlabel("Frequency[pixel^-1]")
    plt.title("f1")
    plt.show()
    #print np.unravel_index(np.argmax(abs(f1)), np.shape(f1))
    #print f0
    #print np.max(f0), np.min(f0), np.argmax(f0), np.argmin(f0)
    print "image shape:", np.shape(f0), "reference shape", np.shape(f1)
    imagereg = np.fft.ifftshift(abs(np.fft.ifft2((f0*np.conj(f1))/(abs(f0)*abs(f1)))))
    print imagereg
    imagereg = scipy.ndimage.gaussian_filter(imagereg, 4)
    plt.plot(freq,np.log10(abs(imagereg[0])), label="0")
    plt.plot(freq,np.log10(abs(imagereg[225])), label="225")
    plt.plot(freq,np.log10(abs(imagereg[512])), label="512")
    plt.plot(freq,np.log10(abs(imagereg[750])), label="750")
    plt.legend()
    #plt.clim(0,0.0003)
    #plt.colorbar()
    plt.show()
    #imagereg = imagereg[512:1536, 512:1536]
    print np.shape(imagereg)
    t0, t1= np.unravel_index(np.argmax(imagereg), np.shape(imagereg))
    t0-=512
    t1-=512
    if t0 > np.shape(imagereg)[0] // 2:
        t0 -= np.shape(imagereg)[0]
    if t1 > np.shape(imagereg)[1] // 2:
        t1 -= np.shape(imagereg)[1]
    print "new shift:", t0,t1
    return (-t1,-t0) #tuple: (y,x) in ds9 image, shift in negative dir. to align with ref



#Borrowed from Christoph Goelke~~~~~~~~~~~~~~~~~~~~~~~~

def similarity(im0, im1):
    """Return similarity transformed image im1 and transformation parameters.

    Transformation parameters are: isotropic scale factor, rotation angle (in
    degrees), and translation vector.

    A similarity transformation is an affine transformation with isotropic
    scale and without shear.

    Limitations:
    Image shapes must be equal and square.
    All image areas must have same scale, rotation, and shift.
    Scale change must be less than 1.8.
    No subpixel precision.

    """
    if im0.shape != im1.shape:
        raise ValueError("Images must have same shapes.")
    elif len(im0.shape) != 2:
        raise ValueError("Images must be 2 dimensional.")

    f0 = np.fft.fftshift(abs(np.fft.fft2(im0)))
    f1 = np.fft.fftshift(abs(np.fft.fft2(im1)))

    h = highpass(f0.shape)
    f0 *= h
    f1 *= h
    del h

    f0, log_base = logpolar(f0)
    f1, log_base = logpolar(f1)

    f0 = np.fft.fft2(f0)
    f1 = np.fft.fft2(f1)
    r0 = abs(f0) * abs(f1)
    ir = abs(np.fft.ifft2((f0 * f1.conjugate()) / r0))
    i0, i1 = np.unravel_index(np.argmax(ir), ir.shape)
    angle = 180.0 * i0 / ir.shape[0]
    scale = log_base ** i1

    if scale > 1.8:
        ir = abs(np.fft.ifft2((f1 * f0.conjugate()) / r0))
        i0, i1 = np.unravel_index(np.argmax(ir), ir.shape)
        angle = -180.0 * i0 / ir.shape[0]
        scale = 1.0 / (log_base ** i1)
        if scale > 1.8:
            raise ValueError("Images are not compatible. Scale change > 1.8")

    if angle < -90.0:
        angle += 180.0
    elif angle > 90.0:
        angle -= 180.0

    im2 = interpolation.zoom(im1, 1.0/scale)
    im2 = interpolation.rotate(im2, angle)

    if im2.shape < im0.shape:
        t = np.zeros_like(im0)
        t[:im2.shape[0], :im2.shape[1]] = im2
        im2 = t
    elif im2.shape > im0.shape:
        im2 = im2[:im0.shape[0], :im0.shape[1]]

    f0 = np.fft.fft2(im0)
    f1 = np.fft.fft2(im2)
    ir = abs(np.fft.ifft2((f0 * f1.conjugate()) / (abs(f0) * abs(f1))))
    t0, t1 = np.unravel_index(np.argmax(ir), ir.shape)

    if t0 > f0.shape[0] // 2:
        t0 -= f0.shape[0]
    if t1 > f0.shape[1] // 2:
        t1 -= f0.shape[1]

    im2 = interpolation.shift(im2, [t0, t1])

    # correct parameters for ndimage's internal processing
    if angle > 0.0:
        d = int((int(im1.shape[1] / scale) * math.sin(math.radians(angle))))
        t0, t1 = t1, d+t0
    elif angle < 0.0:
        d = int((int(im1.shape[0] / scale) * math.sin(math.radians(angle))))
        t0, t1 = d+t1, d+t0
    scale = (im1.shape[1] - 1) / (int(im1.shape[1] / scale) - 1)

    return im2, scale, angle, [-t0, -t1]


def similarity_matrix(scale, angle, vector):
    """Return homogeneous transformation matrix from similarity parameters.

    Transformation parameters are: isotropic scale factor, rotation angle (in
    degrees), and translation vector (of size 2).

    The order of transformations is: scale, rotate, translate.

    """
    S = np.diag([scale, scale, 1.0])
    R = np.identity(3)
    angle = math.radians(angle)
    R[0, 0] = math.cos(angle)
    R[1, 1] = math.cos(angle)
    R[0, 1] = -math.sin(angle)
    R[1, 0] = math.sin(angle)
    T = np.identity(3)
    T[:2, 2] = vector
    return np.dot(T, np.dot(R, S))


def logpolar(image, angles=None, radii=None):
    """Return log-polar transformed image and log base."""
    shape = image.shape
    center = shape[0] / 2, shape[1] / 2
    if angles is None:
        angles = shape[0]
    if radii is None:
        radii = shape[1]
    theta = np.empty((angles, radii), dtype=np.float64)
    theta.T[:] = -np.linspace(0, np.pi, angles, endpoint=False)
    #d = radii
    d = np.hypot(shape[0]-center[0], shape[1]-center[1])
    log_base = 10.0 ** (math.log10(d) / (radii))
    radius = np.empty_like(theta)
    radius[:] = np.power(log_base, np.arange(radii,
                                                   dtype=np.float64)) - 1.0
    x = radius * np.sin(theta) + center[0]
    y = radius * np.cos(theta) + center[1]
    output = np.empty_like(x)
    interpolation.map_coordinates(image, [x, y], output=output)
    return output, log_base


def highpass(shape):
    """Return highpass filter to be multiplied with fourier transform."""
    x = np.outer(
        np.cos(numpy.linspace(-math.pi/2., math.pi/2., shape[0])),
        np.cos(numpy.linspace(-math.pi/2., math.pi/2., shape[1])))
    return (1.0 - x) * (2.0 - x)


#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~


# Necessary steps:

# Build Fourier image/reference aligner: Take FFT of gradient in x and y, Gaussian filter around limb freq, crosscorr, abs val of inverse trans gives shift. Add subpixel func.--Gaussian smooth cross-corr at peak.

# Look up mag corr. in Goodman

# Look up 3-shear rotation

# implement Affine trans: 3-shear, mag, shift at subpixel accuracy.

# Use Subpixel gradient crosscorr to align 1,2 and 4,3 magnetic ratio images.

# Use 3-shear affine trans. to align K, Na images. Add Fe images if things go well.







# Import image data cubes
for i in range(1):#len(Kimnames)):
    image = fits.open(Kimnames[i])
    
    #subtract leak image
    leaksub = np.ndarray(np.shape(image[0].data))
    leaksub = (image[0].data - dark[0].data) - (leakshift - dark[0].data)
    
    #divide by flatfield
    flatdiv = np.ndarray(np.shape(image[0].data))
    flatdiv = leaksub / flat
    
    #plt.imshow(flatdiv[0])
    #plt.show()
    
    ##Get rid of 0s (to prevent NaNs)
    #flatdiv = flatdiv + 0.0001
    
    #shift images 1 and 3 so respective pairs are aligned
    
    #flatdiv[1] = interpolation.shift(flatdiv[1], (20,20))
    flatdiv[2] = flatdiv[2][:,::-1]
    flatdiv[3] = flatdiv[3][:,::-1]
    flatdiv = [interpolation.rotate(flatdiv[x], -71.0, reshape=False) for x in range(4)]
    hdu8 = fits.PrimaryHDU(flatdiv[3])
    hdu8.writeto('Kflatdiv3.fits', clobber=True)
    
    #mask junk around outside of images
    
    for j in range(len(flatdiv[0][0,:])):
        for k in range(len(flatdiv[0][:,0])):
            for l in range(4):
                if flatdiv[l][j,k] < 15000:
                    flatdiv[l][j,k] = 0.0
            #if np.sqrt((len(flatdiv[0][0,:])/2 - j)**2 + (len(flatdiv[0][:,0])/2 - k)**2) > 502:
    
    #measure shift and align images
    plt.imshow(np.log10(flatdiv[0]))
    plt.show()
    plt.imshow(np.log10(flatdiv[2]))
    plt.show()
    similar = similarity(flatdiv[0], flatdiv[2])
    print similar
    
    flatdiv_cent = [image_registration.register_images(flatdiv[0], flatdiv[x],usfac=100) for x in range(4)]
    flatdiv_cent = np.array(flatdiv_cent)
    #flatdiv_cent = np.sqrt(abs(flatdiv_cent))
    print flatdiv_cent
    #print [flatdiv_cent[3,1],flatdiv_cent[3,0]]
    
    #threshhold images to find Sun
    #a = flatdiv > 9000
    #flatdiv_cent = [np.unravel_index(np.argmax(morphology.distance_transform_edt(a[x])),(1024,1024)) for x in range(4)]
    
    #print image[0].shape, flatdiv.shape, flatdiv[0].shape
    imshift1 = interpolation.shift(flatdiv[0], flatdiv_cent[0])# tuple: (y,x) in ds9 image
    imshift2 = interpolation.shift(flatdiv[1], flatdiv_cent[1])
    imshift3 = interpolation.shift(flatdiv[2], flatdiv_cent[2])
    imshift4 = interpolation.shift(flatdiv[3], flatdiv_cent[3])
    
    #plt.imshow(flatdiv[0,:,:]) #flatdiv:[z(imagecube),y,x] in ds9 image
    #plt.show()
    #plt.imshow(flatdiv[0,:,::-1])
    #plt.show()
      
      
    #flip images 3 and 4 along x axis, align with other pair of images
    #imshift3 = interpolation.shift(imshift3[:,::-1], (0.7,-6.5)) #tuple: (y,x) in ds9 image
    #imshift4 = interpolation.shift(flatdiv[3,:,::-1], (0.7,-6.5))
    #imshift3 = imshift3[:,::-1]
    #imshift4 = imshift4[:,::-1]
    
    hdu = fits.PrimaryHDU(imshift2)
    hdu.writeto('Kimshift2.fits', clobber=True)
    #rotate images
    #pad image such that pivot point is at image center and image size is doubled
    #pivot3 = [519,513]
    #pivot4 = [519,508]
    #pad3X = [imshift3.shape[1] - pivot3[0], pivot3[0]]
    #pad3Y = [imshift3.shape[0] - pivot3[1], pivot3[1]]
    #imshift3pad = np.pad(imshift3, [pad3Y,pad3X], 'constant')
    #pad4X = [imshift4.shape[1] - pivot4[0], pivot4[0]]
    #pad4Y = [imshift4.shape[0] - pivot4[1], pivot4[1]]
    #imshift4pad = np.pad(imshift4, [pad4Y,pad4X], 'constant')
    #plt.imshow(imshift3pad)
    
    #rotate images
    #imrot3 = interpolation.rotate(imshift3, 0.0, reshape=False) # switch sign for rotation because x, y switched
    #imrot4 = interpolation.rotate(imshift4, 0.0, reshape=False)
    
    #remove padding and align image centers
    #imrot3 = imrot3[pad3Y[0]:-pad3Y[1], pad3X[0]-6:-pad3X[1]-6]
    #imrot4 = imrot4[pad4Y[0]:-pad4Y[1], pad4X[0]-6:-pad4X[1]-6] #integers are difference between input and output origin
      
    #shift to realign image centers
    #imrot3 = interpolation.shift(imrot3, (7.5,10.0)) #tuple: (y,x) in ds9 image
    #imrot4 = interpolation.shift(imrot4, (7.5,10.0))
    
    #shift to account for relative drift over time
    #imrot3 = interpolation.shift(imrot3, (-3*float(i)/len(Kimnames),-11.0*float(i)/len(Kimnames))) #tuple: (y,x) in ds9 image
    #imrot4 = interpolation.shift(imrot4, (-3*float(i)/len(Kimnames),-11.0*float(i)/len(Kimnames)))
    
    
    #shift images to account for drift over time
    #imshift1 = interpolation.shift(imshift1, (-3.5*float(i)/len(Kimnames),-2.5*float(i)/len(Kimnames))) #tuple: (y,x) in ds9 image
    #flatdiv[1] = interpolation.shift(flatdiv[1], (-3.5*float(i)/len(Kimnames),-2.5*float(i)/len(Kimnames)))
    #imrot3 = interpolation.shift(imrot3, (-3.5*float(i)/len(Kimnames),-2.5*float(i)/len(Kimnames))) #tuple: (y,x) in ds9 image
    #imrot4 = interpolation.shift(imrot4, (-3.5*float(i)/len(Kimnames),-2.5*float(i)/len(Kimnames)))
    
    #plt.imshow(imrot4)
    #plt.show()
    
    hdu2 = fits.PrimaryHDU(imshift3)
    hdu2.writeto('Kimrot3.fits', clobber=True)
    hdu4 = fits.PrimaryHDU(imshift4)
    hdu4.writeto('Kimrot4.fits', clobber=True)
    
    # Generate velocity image from sigma+, sigma- images
    hdu3 = fits.PrimaryHDU(imshift1)
    hdu3.writeto('Kimshift1.fits', clobber=True)
    top = imshift1 - imshift3
    bot = imshift1 + imshift3
    dop1 = top / bot
    top2 = imshift2 - imshift4
    bot2 = imshift2 + imshift4
    dop2 = top2 / bot2
    dop = dop1 + dop2
    # Generate magnetic images
    mag = dop1 - dop2
    mag1 = imshift1 / imshift2
    mag2 = imshift4 / imshift3
    
    #try alinging mag images
    mag2 = interpolation.shift(mag2, image_registration.register_images(mag1, mag2, usfac=100))
    print "calculated mag2 shift:", image_registration.register_images(mag1, mag2, usfac=100)
    plt.imshow(np.log10(mag1))
    plt.show()
    plt.imshow(np.log10(mag2))
    plt.show()
    plt.imshow(np.log10(mag))
    plt.colorbar()
    plt.show()
    
    ##Remove NaNs
    #for j in range(dop.shape[0]):
    #    for k in range(dop.shape[1]):
    #        if numpy.isnan(dop[j,k]) == True:
    #            dop[j,k] = 0.0
    #        elif numpy.isnan(mag[j,k]) == True:
    #            mag[j,k] = 0.0
            #mask edges of images
            #if np.sqrt((j-dop.shape[0]/2.)**2+(k-dop.shape[1]/2.)**2) > 495:
            #    dop[j,k] = 0.0
            #    mag[j,k] = 0.0
    
    #Rotate, flip, unmagnify, shift images to match HMI images
    #dop = interpolation.rotate(dop, 109.0, reshape=False)
    #mag = interpolation.rotate(mag, 109.0, reshape=False)
    #dop = dop[::-1,:]
    #mag = mag[::-1,:]
    #magnif = 0.96
    #dop = interpolation.zoom(dop, magnif)
    #mag = interpolation.zoom(mag, magnif)
    #pad arrays
    #dop = np.pad(dop, ((1-magnif)/2*imrot4.shape[0],), mode='constant', constant_values=(0,))
    #mag = np.pad(mag, ((1-magnif)/2*imrot4.shape[0],), mode='constant', constant_values=(0,))
    #dop = np.pad(dop, (0,1), mode='constant')
    #mag = np.pad(mag, (0,1), mode='constant')
    #dop = interpolation.shift(dop, (5,5))
    #mag = interpolation.shift(mag, (5,5))
    
    ##Shift to align with HMI images
    #dop = dop[:,::-1]
    #mag = mag[:,::-1]
    #magnif = 0.96
    #dopm = interpolation.zoom(dop,magnif)
    #magm = interpolation.zoom(mag,magnif)
    #print "DOPM",np.shape(dopm), "MAGM", np.shape(magm)
    #dopmpad = (1024-dopm.shape[0])/2
    #magmpad = (1024-magm.shape[0])/2
    #print dopmpad,magmpad
    #dop = np.pad(dopm, (dopmpad, dopmpad+1), 'constant') #+1 because odd number of pixels after mag
    #mag = np.pad(magm, (magmpad,magmpad+1), 'constant')
    #dop = interpolation.shift(dop, findcenter(mag, reference))
    #mag = interpolation.shift(mag, findcenter(mag, reference))
    #print np.shape(dop)
    #print np.shape(dop), np.shape(mag)
    
    #plt.clf()
    #plt.imshow(dop)
    #plt.show()
    
    #save as .fits files
    hduv = fits.PrimaryHDU(dop)
    hduv1 = fits.PrimaryHDU(dop1)
    hduv2 = fits.PrimaryHDU(dop2)
    hduv.writeto(Kimnames[i][8:16]+'Kdop.fits', clobber=True)
    hduv1.writeto('Kdop1.fits', clobber=True)
    hduv2.writeto('Kdop2.fits', clobber=True)
    hdum = fits.PrimaryHDU(mag)
    hdum1 = fits.PrimaryHDU(mag1)
    hdum2 = fits.PrimaryHDU(mag2)
    hdum.writeto(Kimnames[i][8:16]+'Kmag.fits', clobber=True)
    hdum1.writeto('Kmag1.fits', clobber=True)
    hdum2.writeto('Kmag2.fits', clobber=True)
    print i
    
tstep = time.time() - t
print("""K image reduction total runtime (in seconds): {0}
        Start time: {1}
""".format(tstep, tstart))
```
