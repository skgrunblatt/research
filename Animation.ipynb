```python
import glob
from astropy.io import fits
import numpy as np
from scipy.interpolate import interp1d


#Plot signal in 2x2 pixel region over time.
Kimnames = sorted(glob.glob('*Kdop.fits'))
Naimnames = sorted(glob.glob('*Nadop.fits'))
Feimnames = sorted(glob.glob('*Fedop.fits'))
print Kimnames[0:3], Naimnames[0:3], Feimnames[0:3]
Nfe = len(Feimnames)
N = len(Kimnames)
Kimlist = np.zeros((N,1024,1024))
Naimlist = np.zeros((N,1024,1024))
Feimlist = np.zeros((Nfe,1024,1024))
Kpixel = np.zeros(N)
Napixel = np.zeros(N)
Fepixel = np.zeros(Nfe)
for i in range(N):
    #print fits.open(imnames[i])[0].data[0]
    Kimlist[i] = fits.open(Kimnames[i])[0].data
    Kpixel[i] = np.sum(Kimlist[i][392:394,488:490])
    Naimlist[i] = fits.open(Naimnames[i])[0].data
    Napixel[i] = np.sum(Naimlist[i][392:394,488:490])
    
#print Kpixel, Napixel

for i in range(Nfe):
    Feimlist[i] = fits.open(Feimnames[i])[0].data
    Fepixel[i] = np.sum(Feimlist[i][392:394,488:490])

#account for linear ramping with least squares fit
tsec = np.linspace(0,len(Kpixel), len(Kpixel))*5
A = np.vstack([tsec,np.ones(len(tsec))]).T
Km, Kc = np.linalg.lstsq(A, Kpixel)[0]
Kpixelcorr = np.zeros_like(Kpixel)
Nam, Nac = np.linalg.lstsq(A, Napixel)[0]
Napixelcorr = np.zeros_like(Napixel)
for i in range(len(Kpixel)):
    Kpixelcorr[i] = Kpixel[i] - (Km*tsec[i]+Kc)
    Napixelcorr[i] = Napixel[i] - (Nam*tsec[i]+Nac)

Kpixel = Kpixelcorr
Napixel = Napixelcorr

tfe = np.linspace(0,len(Fepixel), len(Fepixel))*45 + 12 #account for TAI/UTC difference: 1st pic at 45s - 34s offset = 12s
A = np.vstack([tfe,np.ones(len(tfe))]).T
Fem, Fec = np.linalg.lstsq(A, Fepixel)[0]
Fepixelcorr = np.zeros_like(Fepixel)
for i in range(len(Fepixel)):
    Fepixelcorr[i] = Fepixel[i] - (Fem*tfe[i]+Fec)

Fepixel = Fepixelcorr


plt.plot(tsec, Kpixel)
plt.xlabel('Time (s)')
plt.ylabel('Flux (arbitrary units)')
#plt.title('K 2x2 pixels')
fig = plt.gcf()
fig.savefig('kpixel.png')
plt.show()
plt.plot(tsec, Napixel)
plt.xlabel('Time (s)')
plt.ylabel('Flux (arbitrary units)')
#plt.title('Na 2x2 pixels')
fig = plt.gcf()
fig.savefig('napixel.png')
plt.show()
plt.plot(tfe, Fepixel)
plt.title('Fe 2x2 pixels')
fig = plt.gcf()
fig.savefig('fepixel.png')
plt.show()

#multiply by trig factor at edges to transition smoothly to 0
for i in range(len(Kpixel)):
    if i < 150:
        Kpixel[i] = Kpixel[i] * np.sin(np.pi*i/300)**2
        Napixel[i] = Napixel[i] * np.sin(np.pi*i/300)**2
    elif len(Kpixel)-150 < i < len(Kpixel):
        Kpixel[i] = Kpixel[i] * np.cos(np.pi*(i-(len(Kpixel)-150))/300)
        Napixel[i] = Napixel[i] * np.cos(np.pi*(i-(len(Napixel)-150))/300)
        
for i in range(len(Fepixel)):
    if i < 15:
        Fepixel[i] = Fepixel[i] * np.sin(np.pi*i/30)**2
    elif len(Fepixel)-15 < i < len(Fepixel):
        Fepixel[i] = Fepixel[i] * np.cos(np.pi*(i-(len(Fepixel)-15))/30)
        
        
#Use cubic spline interpolation to make Fepixel array same length
Feinterp = interp1d(tfe, Fepixel, kind='cubic', bounds_error=False, fill_value=0.0)(tsec)
        
plt.plot(tfe,Fepixel)
plt.plot(tsec,Feinterp)
plt.show()
    
#pad arrays
Kpixel = np.pad(Kpixel, (152, 152), 'constant', constant_values=(0,0))
Napixel = np.pad(Napixel, (152, 152), 'constant', constant_values=(0,0))
Fepixel = np.pad(Feinterp, (152, 152), 'constant', constant_values=(0,0))

#FFT
Kpixelfft = np.fft.fft(Kpixel)
Napixelfft = np.fft.fft(Napixel)
Fepixelfft = np.fft.fft(Fepixel)

#Use cubic spline interpolation 
knacompconj = Kpixelfft * np.conj(Napixelfft)
fekcompconj = Fepixelfft * np.conj(Kpixelfft)
fenacompconj = Fepixelfft * np.conj(Fepixelfft)

# make frequency axis
dnu = 1./(len(Pixcompconj)*5.)
print dnu
freq = np.linspace(0,len(knacompconj)*dnu,len(knacompconj))

#Gaussian filter
Gfilter = np.zeros_like(knacompconj)
mu = 0.003 # based one visual inspection of function--how to do this smarter?
sig = 0.0004
print len(freq), len(Gfilter)
for i in range(len(freq)):
    Gfilter[i] = (np.exp(-(freq[i]-mu)**2./(2.*sig**2.))+np.exp(-(freq[i]-((len(Pixcompconj)*dnu)-mu))**2/(2*sig**2)))

knacompconj = knacompconj * Gfilter
fekcompconj = fekcompconj * Gfilter
fenacompconj = fenacompconj * Gfilter

knaifft = np.fft.ifft(knacompconj)
fekifft = np.fft.ifft(fekcompconj)
fenaifft = np.fft.ifft(fenacompconj)

#rearrange ifft arrays to have proper time lags
knaifft = np.append(knaifft[len(knaifft)/2+1:], knaifft[:len(knaifft)/2+1])
fekifft = np.append(fekifft[len(fekifft)/2+1:], fekifft[:len(fekifft)/2+1])
fenaifft = np.append(fenaifft[len(fenaifft)/2+1:], fenaifft[:len(fenaifft)/2+1])

#ADD NEW WINDOW TECH
times = np.linspace(-len(Kpixel)/2,len(Kpixel)/2,len(Kpixel))*5+2.5

plt.plot(times,knaifft) 
plt.xlabel('Time lag (s)')
plt.ylabel('Flux (arbitrary units)')
plt.title('K to Na pixel wavelet')
fig3 = plt.gcf()
fig3.savefig('knapixelwavelet.png')
plt.show()
plt.plot(times,fekifft) 
plt.xlabel('Time lag (s)')
plt.ylabel('Flux (arbitrary units)')
plt.title('Fe to K pixel wavelet')
fig3 = plt.gcf()
fig3.savefig('fekpixelwavelet.png')
plt.show()
plt.plot(times,fenaifft) 
plt.xlabel('Time lag (s)')
plt.ylabel('Flux (arbitrary units)')
plt.title('Fe to Na pixel wavelet')
fig3 = plt.gcf()
fig3.savefig('fenapixelwavelet.png')
plt.show()
#Create mathematical model
#p = A, omega, deltaomega, Tg, Tp
func = lambda p, t: (p[0]**2*(2*np.pi*p[2])/np.sqrt(8*np.pi))*np.exp(-(2*np.pi*p[2])**2*(t - p[3])**2/8)*np.cos(2*np.pi*p[1]*(t-p[4]))
errfunc = lambda p, t, y: func(p,t) - y
p0 = [1000, 0.003, 0.0004, 300, 20] #initial guess of parameters

#Make least squares fit of model to data signals
from scipy.optimize import leastsq
pkna,success = leastsq(errfunc, p0[:], args=(times[len(times)/5:2*len(times)/3], np.real(knaifft[len(times)/5:2*len(times)/3])))

pixtimes = times[len(times)/5:2*len(times)/3]

#plot data and fit together
plt.plot(pixtimes, func(pkna, pixtimes), 'r', pixtimes, knaifft[len(times)/5:2*len(times)/3], 'b')
plt.legend(('model','data'))
plt.xlabel('Time lag (s)')
plt.ylabel('Flux (arbitrary units)')
plt.title('K/Na wavelet fit')
fig = plt.gcf()
fig.savefig('knapixmodel.png')
plt.show()
print("""K/Na pixel wavelet
        A: {0}
        omega: {1} Hz
        deltaomega: {2} Hz
        Tgroup: {3} s
        Tphase: {4} s
        Successful fit? (1 if yes.) {5}
""".format(pkna[0],pkna[1],pkna[2],pkna[3],pkna[4], success))

pfek,success = leastsq(errfunc, p0[:], args=(times[len(times)/5:2*len(times)/3], np.real(fekifft[len(times)/5:2*len(times)/3])))

#plot data and fit together
plt.plot(pixtimes, func(pfek, pixtimes), 'r', pixtimes, fekifft[len(times)/5:2*len(times)/3], 'b')
plt.legend(('model','data'))
plt.xlabel('Time lag (s)')
plt.ylabel('Flux (arbitrary units)')
plt.title('Fe/K wavelet fit')
fig = plt.gcf()
fig.savefig('fekpixmodel.png')
plt.show()
print("""Fe/K pixel wavelet
        A: {0}
        omega: {1} Hz
        deltaomega: {2} Hz
        Tgroup: {3} s
        Tphase: {4} s
        Successful fit? (1 if yes.) {5}
""".format(pfek[0],pfek[1],pfek[2],pfek[3],pfek[4], success))

pfena,success = leastsq(errfunc, p0[:], args=(times[len(times)/5:2*len(times)/3], np.real(fenaifft[len(times)/5:2*len(times)/3])))

#plot data and fit together
plt.plot(pixtimes, func(pfena, pixtimes), 'r', pixtimes, fenaifft[len(times)/5:2*len(times)/3], 'b')
plt.legend(('model','data'))
plt.xlabel('Time lag (s)')
plt.ylabel('Flux (arbitrary units)')
plt.title('Fe/Na wavelet fit')
fig = plt.gcf()
fig.savefig('fenapixmodel.png')
plt.show()
print("""Fe/Na pixel wavelet
        A: {0}
        omega: {1} Hz
        deltaomega: {2} Hz
        Tgroup: {3} s
        Tphase: {4} s
        Successful fit? (1 if yes.) {5}
""".format(pfena[0],pfena[1],pfena[2],pfena[3],pfena[4], success))
```
