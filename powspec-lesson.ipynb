# Scientific Coding in Python: Creating a Power Spectrum from Image Data

Consider the following problem: You have a sequence of images of an oscillating surface. These oscillations occur both on various timescales and lengthscales on the surface of interest. How can we decompose the complex motion of the surface in order to determine the different frequencies and wave lengths that are strongest in the object of interest?

For this lesson, we will use images of the Sun, the surface of which can be thought of as a fluid that is constantly oscillating. These images have been taken every 10 seconds for approximately 8 hours, and we will use this time-series to attempt to uncover the frequencies of interest that are present. Once we establish how to apply these techniques to our images, we'll discuss how these techniques can be used in other situations, and where else such techniques could come in handy.

## Relevant Background Vocabulary

- Power: The relative strength of a signal, usually at a given frequency or range of frequencies.
- Power Spectrum: The relative spread of power across the full range of frequencies.
- Wave Number `k`: `k = 2pi/lambda, lambda = Nx * Deltax`. This will make more sense once we've taken the Fourier transform of our data. 
- Fourier Transform: A mathematical transformation of data given in Euclidean (regular) space to frequency space, used to identify the strongest frequencies and wavelengths of oscillation in a physical dataset. 
- Complex Conjugate: Consider a complex number (with both a real and imaginary part). The complex conjugate of this number is the same number, except with the sign of the imaginary part reversed. 
- Azimuthal Average: An average over all points a certain distance from the axis of interest. In this notebook, we azimuthally average over the spatial dimensions (by adding their averages in quadrature) to allow us to represent our three dimensional data in two dimensions.
- Adding in Quadrature: Square root of the sum of two squared numbers: `c = sqrt(a^2 + b^2)`

For this project, we're trying to identify the features of the power at all wave lengths (wave numbers) and frequencies by creating a power spectrum. As our data is three-dimensional, the power spectrum will also be three dimensional, and so we explore different ways to present it in two dimensions in order to quickly decipher the important information.

# Importing Data

Next, we need to import the relevant dataset. As this is an astronomical dataset, it has been saved as an IDL .sav file, which contains a header and then the entire stack of relevant data. Luckily, Python has a built-in library to read in .sav files, and so accessing the relevant data is straightforward.
```python
Kdata = scipy.io.readsav('short_k_v_stack.sav',python_dict=True) # Read in data, and save as Python dictionary.
Kimlist = Kdata['stack_v_k'] # Define the data stack.
#Kt = Kdata['times_v_k'] # Should we show people where the time cadence came from?
```

## Plotting our Data

Let's take a moment to get to know the data. First, we should make sure we have an understanding of the size and shape of our images. 

```python
plt.imshow(Kimlist[100], cmap='cubehelix') # Specify color map to make for easier visual understanding.
plt.show()
```

Note that the `imshow` function built into `matplotlib` will plot the array indices as axis labels automatically. This gives us a quick and dirty way to be familiar with the data we're working with--we can see that we have approximately 180 pixels in the x-direction and 120 pixels in the y-direction.

A gradient is visible across the image. We would expect this, as we're effectively measuring line-of-sight velocities in these images, and thus the rotation of the Sun would result in such a pattern. Variations due to additional motions within the Sun would break this pattern. We can see an example of this in the bottom center section of the image. Cool!

Next, we can plot the variation in one pixel over time. This will give us a quick insight into the timescale of the strongest oscillations present in the data.

```python
plt.plot(Kimlist[:1000,50,50]) #Plot slice through Python array corresponding to one spatial location in the first 1000 images.
plt.xlabel('Image (10s cadence)')
plt.ylabel('Signal')
plt.show()
```

We can already see multiple components of the signal here. There seems to be a high frequency oscillation as well as larger trend changes over the datastring. Now, we want to separate these components of the signal from one another.

Now that we understand the shape of the Python lists that are holding the data of interest, we can pass the data to the relevant built-in Python function, an n-dimensional Fourier transform. This will change the axes of the data from (length, length, time) to (1/length, 1/length, 1/time). This allows us to explore the data as a function of temporal and spatial frequency, giving us more direct information about the features of the oscillations present, as opposed to showing us the signal created by one individual location in space or time.

```python
#Create a power spectrum.

#First, take n-dimensional Fourier Transform of image data.
Kfftspec=np.fft.fftn(Kimlist, s=(4096,256,256)) # FFT
Kfftspec = np.fft.fftshift(Kfftspec) # Shift FFT such that lowest frequencies are at the left of the array, highest at the right.

#Power Spectrum = FFT * conj(FFT)
Kpowspec = Kfftspec * np.conj(Kfftspec)
t = time.time() - tstart
print "FFT runtime:", t, "s"
```

**Note the shape of the FFT array.** We have added buffer area outside the original image pixels in the spatial dimensions to allow for a smoother FFT. We should do this in the time dimension as well, but since we have such a long time series relative to the frequencies of interest, it is less necessary.

**Also, we take the complex conjugate of the FFT to give us the power spectrum.** This allows us to combine the real and imaginary parts of the FFT so we can measure absolute power instead of power in the real or imaginary direction. 

## Representing the Power Spectrum

Great! We've made a power spectrum of our data. Now, all we have to do is reconfigure it so that it's easy to interpret.

First, instead of plotting both positive and negative frequencies separately, let's align them on top of one another, as they should be mirror images of one another for this dataset. This is effectively folding the data perpendicular to the time axis.

```python
#Analysis: run separately from FFT
tstart = time.time()

#combine positive and negative frequencies
newKpowspec = np.zeros((2048,256,256))
for i in range(len(Kpowspec)/2):
    newKpowspec[i] = Kpowspec[i] + Kpowspec[-(i+1)] # combine Kpowspec[1]
    
newKpowspec = newKpowspec[::-1] #flip over frequency axis so frequencies run from lowest to highest.
```

Let's make a plot of the data at a given time frequency to make sure that we've arranged the data in a reasonable way.

```python
plt.imshow(np.log((newKpowspec[180]))) #We plot the logarithm of the power to make the features more visible.
plt.show()
```

In this plot, we can identify the typical oscillations in power as a function of spatial frequency, or wavenumber. Great! However, we're most interest in comparing the power of spatial oscillations to the power of temporal oscillations. Thus, we need to find a way to collapse this two-dimensional spatial information into one dimension. 

As you can see here, the spatial information seems to be tied to the distance from the origin--it's circular in the above representation. Thus, it would make sense to represent the two-dimensional information as a function of the distance from the origin. 

We handle this by finding the average value for a given distance from the origin for each frequency. This is referred to as *azimuthally averaging* the data. 

```python
kaz = np.zeros((2048,256))
#compute azimuthal avg
for i in range(256):
    for j in range(2048): # loop over all index locations in kaz
        kx = np.array(newKpowspec[j,i,:]) # relevant slice along x axis
        ky = np.array(newKpowspec[j,:,i]) # slice along y axis
        kaz[j,i] = np.sqrt(np.mean(kx)**2 + np.mean(ky)**2) #azimuthal average = x and y averages added in quadrature.
```

Let's take a look at our `kaz` array to make sure we're on the right track.

```python
plt.imshow(np.log(kaz))
plt.show()
```

We can definitely see some structure in this plot! Now, we stack the positive and negative components, although now in spatial frequency, to make the data more readable.

```python
newkaz = np.zeros((2048,128))
for i in range(len(kaz[0,:])/2):
    for j in range(2048):
        newkaz[j,i] = kaz[j,i] + kaz[j,-(i+1)] #combine positive and negative spatial values for each temporal frequency. 
newkaz[:,:] = newkaz[:,::-1] # flip over spatial frequency axis.
```

Now, let's plot the data. We should also label our axes, which is not a trivial task. In order to do so, we need to compute the full range of spatial and temporal frequencies given in our power spectrum, and then scale the axis labels accordingly in frequency and wavenumber units. We do this below:

```python
Nt = 4096 #steps in temporal direction
Nx = 256 # steps in spatial x direction
Ny = 256 # steps in spatial y direction
deltat = 10 #s
deltax = 3.86 #Mm/pixel. Mm: megameter = 1 million meters. 
deltay= 3.86 #Mm/pixel
dt = round(float(1. / (Nt*deltat)),6) # step size in temporal direction, rounded to the nearest hundredth
dnux = round(float(2*np.pi / (Nx*deltax)),6) #step size in spatial direction. multiply by 2pi factor due to 
                                             # wavenumber definition: k=2pi/lambda, lambda=Nx*deltax. round to make plot labels legible.
dnuy = round(float(2*np.pi / (Ny*deltay)),6)

print round(dt,6), dnux, dnuy
dnu = np.sqrt(dnux**2+dnuy**2)
dnu = round(dnu,4)
print dnu
```

We can use these definitions when defining our plot axis labels.

```python
plt.imshow(np.log(newkaz[:250,:]), cmap='cubehelix', interpolation='none') # trim longer frequencies for easier viewing.
plt.ylabel('Frequency (mHz)')
ylocs = [0,30,60,90,120,150,180,210,240] # locations of axis labels, in pixels
dt = dt *1000 #milliHertz (mHz) instead of Hertz. 1 Hertz = 1 oscillation per second.
yvals = [0,30*dt,60*dt,90*dt,120*dt,150*dt,180*dt,210*dt,240*dt] # values of axis labels
plt.yticks(ylocs,yvals) #places axis labels equal to yvals at ylocs pixel locations in plot.
plt.xlabel('Wavenumber (Mm^-1)')
xlocs = [0,30,60,90,120]
xvals = [0,30*dnu, 60*dnu, 90*dnu, 120*dnu]
plt.xticks(xlocs,xvals)
plt.show()
```

And now we have our easily readable power spectrum! 
- The strongest values are seen at low frequency and low wavenumber. This is an artifact of having finite data, and we can ignore it. The plaid-like pattern in the bottom left hand corner is also due to this effect.
- The yellowish patch around 3mHz at low wavenumbers corresponds to the fundamental frequency of solar oscillations. This corresponds to the high frequency sine curve we illustrated for a given pixel earlier, but since it is present in each pixel, it gives us little spatial information. These waves can be thought of as similar to waves on the surface of the ocean. 
- The ridges to the right of this, however, correspond to the pressure modes of oscillation, giving us more information about the internal structure of the object. Isolating these oscillations allows us to probe underneath the surface of whatever we're looking at.

We can also display the power spectrum for a given wavenumber to look at the strength of oscillations of a given spatial size as a function of frequency. 

```python
plt.plot(np.log(kaz[:,0]))
plt.plot(np.log(kaz[:,10]))
plt.plot(np.log(kaz[:,100])) # these correspond to different wavenumbers: waves of different spatial sizes.
plt.legend(('kaz0','kaz10', 'kaz100'))
plt.show()
print "Max value and location:",np.max(kaz), "at", np.unravel_index(np.argmax(kaz), np.shape(kaz))
```

This frequency analysis can then be used to tell us a great deal about the internal structure and dynamics of stars. At all wavenumbers, the 3mHz peak is visible. At smaller wavenumbers, however, there appears to be structure within the 3 mHz peak, and at the larger wavenumber (`kaz100`) some weaker structure can be seen at larger frequencies. 

The strongest frequency and spacing between frequencies is directly dependent on the density of the oscillating material. We use this to infer the stage of stellar evolution of the star, and can use this information for more precise constraints on stellar density, age, and therefore temperature, size, and mass.

This type of cross-correlation frequency analysis can be applied to any data where one wants to use surface wave oscillations to learn more about what lies underneath. For example, analysis of standing waves on a water surface can tell you more about the depth and shape of the underlying body of water.
