## Background

Consider Newton's discovery of the force of gravity. It is *possible* that he did immediately come up with the equation for the gravitational force when the apple woke him from his nap, as the legend says. But it seems much more likely that he had to test many different equations (AKA theoretical models) to determine one relation that would describe how an apple falls to the ground and also how the Moon stays in the sky. Before choosing the famous equation of gravity we all know today, $F_{g} = \frac{GM_1M_2}{r^2}$, he probably tried countless other equations and parameter values, but then tested how well they described the motion of objects falling on Earth as well as the orbit of the Moon around the Earth, and found that the particular equation that described the motion best was the one above. 

Similarly, scientists today produce many tests of whatever they're studying, which they call observations. One of the most essential parts of any scientific study is using a theoretical model, such as (but not necessarily!) an equation, to describe those observations. In order to find a good model, scientists must compare many models to the data and then decide which one describes the data most accurately. 

In this lesson, we're going to learn how to apply the simplest method for comparing models that scientists use regularly - the **least squares** method - to analyze artificial and then real data in Python. In this method, we use one equation with various parameters to describe our observations. We then test thousands of parameter values to find those that describe the observations most precisely, by calculating the difference between the model and the observations for each particular set of parameters. This method can also be used to compare between different equations with various parameters. However, in this lesson, we're going to skim over the details of how this works (because it's already been explained well [here](http://www.physics.csbsju.edu/stats/least_squares.html), [here](http://mathworld.wolfram.com/LeastSquaresFitting.html) and [here](https://en.wikipedia.org/wiki/Least_squares)) and instead focus on how this type of test is implemented in Python.

## Vocabulary

- **Wavelet**: A wave-like oscillation with an amplitude that begins at zero, increases, and then decreases back to zero. Wavelets can be described uniquely by the Wavelet Equation: $y(t) = \frac{A^2\delta\omega}{\sqrt{8\pi}}\mathrm{exp}\Bigg[-\frac{\delta\omega^2(t-T_g)^2}{8}\Bigg]\mathrm{cos}[\omega(t-T_p)]$, whose parameter values we'll discuss later.
- **Least Squares**: A method of fitting a model to data so as to minimize the sum of the squares of the differences between the observed values and the estimated values.
- **Error Function**: Function that returns the difference between the model function value and the measured value.
- **Covariance Matrix**: A matrix describing the least squares difference between each model function value and measured value, used to evaluate the quality of the model fit.

### Create Arbitrary Function

Next, we need to artificially create our data. Eventually, we'd like to use this on real data, but that's a bit more complicated and will be discussed in the next lesson. In this case, we're going to make a *wavelet*. 

<img src="https://upload.wikimedia.org/wikipedia/commons/4/45/Two-frequency_beats_of_a_non-dispersive_transverse_wave_(animated).gif">

In the animation above, individual wavelets can be thought of as the wave signal between two adjacent green dots. Each wavelet can be described by five different parameters: an amplitude, wavelength, spread in wavelength, and group and phase times. In the above animation, all wavelets pictured would have the same parameter values, with the exception of the group time, because all the wavelets have the same size and shape, and are in phase with one another. Therefore, the only difference between them is the time at which they hit their peak signal value, which is governed by the group time.

Mathematically speaking, a wave is described by a sine curve. To make a wavelet, this sine curve must be convolved with a Gaussian function, giving it the 'energy packet' form that we recognize as a wavelet. The equation for a wavelet signal as a function of time is $y(t) = \frac{A^2\delta\omega}{\sqrt{8\pi}}\mathrm{exp}\Bigg[-\frac{\delta\omega^2(t-T_g)^2}{8}\Bigg]\mathrm{cos}[\omega(t-T_p)]$, where $T_p, T_g, A, \omega,$ and $\delta\omega$ are variables describing the phase time, group time, amplitude, $2\pi \times$frequency, and $2\pi \times$frequency width, respectively. The amplitude is the maximum of the wave signal, and the frequency and frequency width describe the shape of the individual oscillations and the total width of the wavelet in time, respectively. The group time corresponds to the peak of the entire wave packet, whereas the phase time corresponds to the location of individual oscillations (and thus will not be larger than the wavelength, 1/frequency).

Below, we've written the equation for a wavelet as `leastsq_func` and `curve_fit_func`. In both cases, the equation takes five parameters and a time variable. We've chosen to use both types of function definitions to show how either can be used for this model fitting case. For more information on how the `lambda` function works, check it out [here](http://www.secnetix.de/olli/Python/lambda_functions.hawk).

```python
leastsq_func = lambda p, t: (p[0]**2*(2*np.pi*p[2])/np.sqrt(8*np.pi))*np.exp(-(2*np.pi*p[2])**2*(t - p[3])**2/8)*np.cos(2*np.pi*p[1]*(t-p[4]))

def curve_fit_func(t,p0,p1,p2,p3,p4):
    return (p0**2*(2*np.pi*p2)/np.sqrt(8*np.pi))*np.exp(-(2*np.pi*p2)**2*(t - p3)**2/8)*np.cos(2*np.pi*p1*(t-p4))

#Test fit to model, see what params are returned
testimes = np.linspace(-3600,3600,720)
A = 1000
omega
testwave = (1000**2*(2*np.pi*0.0008)/np.sqrt(8*np.pi))*np.exp(-((2*np.pi*0.0008)**2*(testimes - 1000)**2)/8)*np.cos(2*np.pi*0.003*(testimes-500))

#Plot the test wavelet.
plt.plot(testimes,testwave)
plt.show()
```

### `leastsq` vs. `curve_fit`

Since there are (at least) two packages in `scipy.optimize` that can be used for least squares fitting, `leastsq` and `curve_fit`, we've included both of them here. The main difference between the two is that `curve_fit` accepts our model, `curve_fit_func`, directly as an argument, whereas `leastsq` requires an additional step to create an error function, `errfunc`, which outputs the *difference* between the data and the model. Both will output the best fit set of parameters after testing a wide range of parameter values, **defined by the initial set of parameters that we provide**. This caveat is a large weakness of the least squares method, as in order to find an accurate set of parameters, one must first have a good idea of what they're close to. This is often undesirable in the case in scientific studies, but still provides a very valuable starting point in many cases.

```python
p0 = [1000, 0.004, 0.01, 1000, 450] #initial guess of parameters as a Python list

errfunc = lambda p, t, y: leastsq_func(p,t) - y

#Make least squares fit of model to fake data signals
p1,p1cov = leastsq(errfunc, p0, args=(testimes, testwave)) # returns optimized parameters in p1, covariance in p1cov.
p2,p2cov = curve_fit(curve_fit_func, testimes, testwave, p0)
```

### Plot Result

Ok, now that we've calculated the best fit parameter values using both the `leastsq` and `curve_fit` packages, we can compare both models to the actual data by plotting them altogether. If the models are good, we should only see one line, as the data and model would all be on top of one another, and the parameter values should match. We'll also plot the models with the initial parameters to ensure that the fitting algorithms have changed things. We also look at the sum of the covariance matrix, which describes the relationship between each data point and corresponding model value, which should be small if the fit was successful.

```python
#plot initial fit guesses
plt.plot(testimes, leastsq_func(p0,testimes), 'r', testimes, curve_fit_func(testimes,p0[0],p0[1],p0[2],p0[3],p0[4]), 'g')
plt.legend(('init. LS model','init. CF model'))
plt.show()

#plot data and fit together
plt.plot(testimes, leastsq_func(p1, testimes), 'r', testimes, testwave, 'b', \
         testimes, curve_fit_func(testimes,p2[0],p2[1],p2[2],p2[3],p2[4]), 'g')
plt.legend(('data','LS model','CF model'))
plt.title('Wavelet fit')
plt.show()
print("""Wavelet parameter values:
        Amplitude: {0} (LS), {6} (CF)
        Frequency: {1} Hz (LS), {7} Hz (CF)
        Frequency spread: {2} Hz (LS), {8} Hz (CF)
        Group time: {3} s (LS), {9} s (CF)
        Phase time: {4} s (LS), {10} s (CF)
        Sum of Covariance Matrices (should be less than infinity): {5} and {11}
""".format(p1[0],p1[1],p1[2],p1[3],p1[4], np.sum(p1cov), p2[0],p2[1],p2[2],p2[3],p2[4], np.sum(p2cov)))
```
